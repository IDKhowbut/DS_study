```python
# 히스토그램
'''
    x변수가 가질 수 있는 값의 구간
    값만 필요하고 인덱스는 필요없다.
'''
```

```python
s3 = pd.Series(np.random.normal(0, 1, size=100))   # 정규분포표에서 뽑는 난수
s3.head()
```

```python
s3.hist()   # 구간 10개로 잡아주는게 디폴트?
s3.hist(bins=50)
```

```python
x1 = np.random.normal(1, 1, size=(100, 1))
x2 = np.random.normal(-2, 4, size=(100,1))

x = np.concatenate((x1, x1), axis=1)  # 축의 기준: 행 으로 위아래 붙이기

# dataframe의 head기능이 numpy에서는 뭘로 쓰이게? => slicing
x[:5]   # 왼쪽에 x1, 오른쪽에 x2
```

```python
df3 = pd.DataFrame(x, columns=["x1", "x2"])
df3.head()  # dataframe으로 만들면 head()기능 사용이 가능!
```

```python
# 산포도는 변수 2개여야함  (회귀분석에서 단골처럼 사용됌)

plt.scatter(df3["x1"], df3["x2"])
```
