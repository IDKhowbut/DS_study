# Series

```python
import numpy as np
import pandas as pd
```

```python
s=pd.Series([9904312, 3448737, 2890451, 2466052])  # list..
s    # enumerate 비슷?
# print(type(s))
```

```python
for p in enumerate([9904312, 3448737, 2890451, 2466052]):
    print(p)
```

```python
# 2015년 인구
s=pd.Series([9904312, 3448737, 2890451, 2466052], index=["서울", "부산", "인천", "대구"])
s
s.values  # 배열 타입임을 알 수 있음
s.index

s.name = "인구"
s
s.index.name = "도시"   # 데이터 틀로서도 충분히 활용가능
s
```

```python
# 연산
s1 = s/1000000   # 값에만 영향
s1
```

```python
# 인덱싱
print(s[1], s["부산"])  # 문자로 하는 인덱싱도 가능하다!
print(s[[0,3,1]], s[["서울", "대구", "부산"]])
print(s[(250e4<s) & (s<500e4)])
print()
print(s[1:3])           # 부산, 인천까지 print
print(s["부산":"대구"])  # 문자로 slicing할때에는 끝설정까지 나온다 = 대구까지 나온단 소리임
print()
print(s.부산, ",", s.서울)
```

```python
# series와 dict
print("서울" in s)   # True
print("대전" in s)   # False

for k, v in s.items():   # 키와 값을 따로 분리
    print("%s = %d" %(k,v))
    
# 2010년도 인구에 대한 데이터
s2 = pd.Series({"서울":9631482, "부산":3393191, "인천":2632035, "대전":1490158}, 
               index=["인천", "대전", "서울","부산"])   # dict=순서없는 자료형, index[]로 별도로 순서를 정할 수 있음
s2
```

```python
# 인덱스 기반 연산  # 별도 지정이 없을 때에는 index를 기반으로 연산한다.

ds = s - s2    # 값없을때 NaN = 숫자가 없다, 데이터가 없다라는 뜻
ds

print(s.values - s2.values)   # values만 이용

ds.notnull()  # 값이 없으면 False로 나온다
ds[ds.notnull()]  # True값만 넘어감?

rs = (s-s2)/s2 * 100    # 인구증가율 확인가능
rs = rs[rs.notnull()]   # 의미있는 데이터만 가져오기?
rs
```

```python
# 데이터의 추가, 갱신, 삭제
rs["부산"] = 1.63
rs

rs["대구"] = 1.41    # 없으면 새로 만들어짐
rs

del rs["서울"]
rs
```

# DataFrame

```python
data = {
    "2015": [9904312, 3448737, 2890451, 2466052],
    "2010": [9631482, 3393191, 2632035, 2431774],
    "2005": [9762546, 3512547, 2517680, 2456016],
    "2000": [9853972, 3655437, 2466338, 2473990],
    "지역": ["수도권", "경상권", "수도권", "경상권"],
    "2010-2015 증가율": [0.0283, 0.0163, 0.0982, 0.0141]
}
print(type(data))  # dict

df = pd.DataFrame(data, columns=["지역", "2015", "2010", "2005", "2000","2010-2015 증가율"], 
                 index=["서울", "부산", "인천", "대구"])   # column 순서 지정
print(type(df))    # Dataframe
df    # 열 우선기준
```

```python
df.values
df.columns
df.index

df.index.name = "도시"
df.columns.name = "특성"
df

df.T    # 행과 열의 위치 바꾸기(Transpose)
```

```python
# 열 데이터의 갱신, 추가, 삭제
print(type(df["2010-2015 증가율"]))    # 열 위치, type:Series 형태
df["2010-2015 증가율"] = (df["2010-2015 증가율"] * 100).round(2)
df

df["2005-2010 증가율"] = ((df["2010"] - df["2005"])/df["2005"] * 100).round(2)
df

del df["2010-2015 증가율"]
df
```

```python
# 열 인덱싱
df["지역"]     # series
df[["지역"]]   # series에서 []로 한번 더 묶어주면 DataFrame

df[["2010", "2005"]]

# df[0]    # column이 문자로 지정되었을 때에는 정수형으로 쓸 수 없다
```

```python
# 행 인덱싱(반드시 : 으로 슬라이싱)
df[:1]    # 첫번째 행만 뽑아내기   열이 아니라 행이다! 헷갈리지말것(numpy랑 헷갈릴듯)
df[1:2]   # 부산만 나옴 = 첫번째 행..
df["서울":"부산"]
```

```python
# 개별 데이터 인덱싱
df["2015"]["서울"]
```
