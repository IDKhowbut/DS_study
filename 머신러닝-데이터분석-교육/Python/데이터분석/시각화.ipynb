```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
# plt 이용 (도화지에 그래프 그리기)
fig = plt.figure()  # 도화지 준비
ax = fig.add_subplot(1, 1, 1)  # 어느 위치에 그래프 그릴건지 1x1그리드(행,열)에 첫 번째 subplot  2행 2열이면 그래프 4개를 그리겠다는 뜻
# 234 : 2x3그리드에 네 번째 subplot / nml : nxm그리그에 l번째 subplot

ax.hist(tips["total_bill"], bins=20) # 기본 단위값: 10, 조정은 bins으로 한다(y축..)
ax.set_title("Histogram of Total Bill")
ax.set_xlabel("frequency")
ax.set_ylabel("Total")
```

```python
# 이변량 -> 산포도 (두 변수 비교)
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(tips["total_bill"],tips["tip"])
ax.set_title("Scatterplot of Total Bill vs Tip")
ax.set_xlabel("Total_bill")
ax.set_ylabel("Tip")
```

```python
# Boxplot: 다른 형식의 데이터를 하나의 그래프로 그릴 수 있음
# 성별, tip = 질적, 양적

fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.boxplot([tips[tips["sex"]=="Female"]["tip"], tips[tips["sex"] == "Male"]["tip"]], labels=["Female", "Male"])
ax.set_title("Boxplot of Tips by Sex")
ax.set_xlabel("Sex")
ax.set_ylabel("Tip")

# Male이 좀 더 퍼져있음 -> size가 더 커서 나오는 현상일 수 있다. 대체로 sex에 따른 tip의 변화는 심하지 않다고 볼 수 있다?
```

```python
# 성별을 숫자로 만들기 위한 함수 만들기
def recode_sex(sex):
    if sex == "Female":
        return 0
    else:
        return 1

# tips의 sex열을 불러와서 recode_sex함수를 적용시켜 성별을 숫자로 치환한다
tips["sex_color"] = tips["sex"].apply(recode_sex)
tips.head()
```

#### scatter
    matplotlib.pyplot.scatter(x, y, s=None, c=None, marker=None, cmap=None, norm=None, vmin=None, vmax=None, alpha=None, 
    linewidths=None, verts=None, edgecolors=None, *, data=None, **kwargs)

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.scatter(x=tips["total_bill"], y=tips["tip"], c=tips["sex_color"], s=tips["size"]*10, alpha=0.5)  # size: 동그라미들 크기지정
ax.set_title("Scatterplot of Total Bill vs Tip")
ax.set_xlabel("Total_bill")
ax.set_ylabel("Tip")
```

```python
import seaborn as sns
tips = sns.load_dataset("tips")
```

```python
ax = plt.subplots()   # 그래프 도화지 만들기

#히스토그램을 seaborn으로 그리면 어떻게 달라지나 살펴보기
ax = sns.distplot(tips["total_bill"])  # 그래프의 포물선을 밀집도(density)라고 부른다.
ax.set_title("Total Bill")
ax.set_xlabel("Total_bill")
ax.set_ylabel("Frequency")
```

```python
ax = plt.subplots()   # 그래프 도화지 만들기
ax = sns.distplot(tips["total_bill"], kde=False)    # kde: 커널 밀도 특정 (kde=False, 포물선 빼기)
ax.set_title("Total Bill")
ax.set_xlabel("Total_bill")
ax.set_ylabel("Frequency")
```

```python
ax = plt.subplots()   # 그래프 도화지 만들기
ax = sns.distplot(tips["total_bill"], hist=False)    # hist=False : histogram 빼기
ax.set_title("Total Bill")
ax.set_xlabel("Total_bill")
ax.set_ylabel("Frequency")
```

```python
ax = plt.subplots()   # 그래프 도화지 만들기
ax = sns.distplot(tips["total_bill"], rug=True)  # 양탄자그래프(rug) 추가 : frequency 높을수록 양탄자가 촘촘하고 진하게 보임
ax.set_title("Total Bill")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Frequency")
```

```python
ax = sns.countplot("day", data=tips)
```

```python
# 회귀선
# fit_reg 옵션으로 회귀선을 제거할 수 있다.
ax = sns.regplot(x = "total_bill", y="tip", data=tips, fit_reg=False)
```

```python
# 2차원 실수형 데이터
# 스캐터 플롯뿐 아니라 차트의 가장자리(margin)에 각 변수의 히스토그램도 그린다.
ax = sns.jointplot(x = "total_bill", y="tip", data=tips)
```

```python
ax = sns.jointplot(x = "total_bill", y="tip", data=tips, kind="hex")  # 점의 모양을 육각형으로(kind="hex"), 데이터가 너무 많으면 별로인 방법
```

```python
# 밀집도plot
ax = sns.kdeplot(data=tips["total_bill"], data2=tips["tip"], shade=True)
```

```python
# 시간대별 변화? 움직임? 보기 (barplot으로)
ax = sns.barplot(x="time", y="total_bill", data=tips)
```

```python
# boxplot은 다양한 정보를 한번에 확인해준다는 장점이 있다.
# 중앙값, 표준 편차 등, 분포의 간략한 특성만 보여줌
ax = sns.boxplot(x="time", y="total_bill", data=tips)
```

```python
# 바이올린 플롯 (http://seaborn.pydata.org/generated/seaborn.violinplot.html)
# 카테고리값에 따른 각 분포의 실제 데이터나 전체 형상을 보여준다
# 이 플롯으로 분산의 정도가 대략적으로 확인이 가능하다
ax = sns.violinplot(x="time", y="total_bill", data=tips)
```

```python
# 다차원 실수형 데이터
# pairplot( http://seaborn.pydata.org/generated/seaborn.pairplot.html)
# 데이터프레임을 인수로 받아 그리드(grid) 형태로 각 데이터 열의 조합에 대해 스캐터 플롯을 그린다. 
# 같은 데이터가 만나는 대각선 영역에는 해당 데이터의 히스토그램을 그린다.
ax = sns.pairplot(tips)
```

```python
ax = sns.PairGrid(tips)
ax.map_upper(sns.regplot)
ax.map_lower(sns.kdeplot)
ax.map_diag(sns.distplot, rug=True)
```

```python
ax = sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", fit_reg=False)
```

```python
ax = sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", fit_reg=False,
               scatter_kws={"s":tips["size"]*10})
```

```python
# markers로 카테고리 구분을 할 수 있다!
ax = sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", fit_reg=False,
               scatter_kws={"s":tips["size"]*10}, markers=["o", "x"])
```
