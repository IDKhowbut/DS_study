### # 파이썬 사이킷런으로 나이브 베이즈 분류기 문제 풀어보기!
    - 문일철 교수님 기계학습 CH3.의 퀴즈

```python
import pandas as pd
import numpy as np
```

```python
# 주어진 데이터
df = pd.DataFrame({'Sky':['Sunny','Sunny', 'Rainy', 'Sunny'],
                   'Temp':['Warm','Warm','Cold','Warm'],
                 'Humid':['Normal', 'High', 'High', 'High'],
                 'Wind':['Strong','Strong', 'Strong', 'Strong'],
                 'Water':['Warm', 'Warm', 'Warm', 'Cool'],
                 'Forecst':['Same','Same','Change','Change'],
                 'EnjoySpt':['Y','Y', 'N', 'Y']})
df
```

```python
# 변수값을 수치화한다.
# Import LabelEncoder (string -> int)
from sklearn import preprocessing

#creating labelEncoder
le = preprocessing.LabelEncoder()

# (string -> int)
print('인코딩할 컬럼들: ', df.columns)
for col in df.columns.tolist():
    print(col)
    name = f'{col}_encoded'
    df[name] = le.fit_transform(df[col].values)
```

```python
display(df)
print(df.columns)
# training 에 사용할 칼럼들
x_columns = ['Sky_encoded', 'Temp_encoded', 'Humid_encoded', 'Wind_encoded','Water_encoded', 'Forecst_encoded']
```

```python
df[x_columns].values
```

```python
df['EnjoySpt_encoded'].values
```

### #사이킷런에서 제공하는 나이브베이즈 모형
    사이킷런의 naive_bayes 서브패키지에서는 다음과 같은 세가지 나이브베이즈 모형 클래스를 제공한다.

- GaussianNB: 정규분포 나이브베이즈
- BernoulliNB: 베르누이분포 나이브베이즈
- MultinomialNB: 다항분포 나이브베이즈

```python
# 모형 3개를 하나씩 적용하여 그 결과를 비교해보자!
# training and test 과정
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB, MultinomialNB
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Split dataset into training set and test set
# 70% training and 30% test
X_train, X_test, y_train, y_test = train_test_split(df[x_columns].values, df['EnjoySpt_encoded'].values, test_size=0.3, random_state=109)

#Create a Classifier
#clf = GaussianNB()  # 가우시안 분포 나이브 베이즈
#clf = BernoulliNB() # 베르누이 분포 나이브 베이즈
clf = MultinomialNB() # 다항분포 나이브 베이즈

# Train the model using the training sets
clf.fit(X_train, y_train)
# predict test set
y_pred = clf.predict(X_test)
# Model Accuracy, how often is the classifier correct? 분류기 정확도 측정
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```

### # 예측할 데이터 만들기

```python
# 실제 예측할 데이터의 변수 : {'Sky':'Sunny', 'Temp':'Warm', 'Humid':'High', 'Wind':'strong', 'Water':'Warm', 'Forecst':'Change'}
# 제대로 된 라벨인코딩을 위해 더미를 넣어줌
new = pd.DataFrame({'Sky':['Sunny', 'Rainy', 'Sunny', 'Rainy'],
                        'Temp':['Warm', 'Cold', 'Warm', 'Cold'],
                         'Humid':['High', 'Normal', 'High', 'High'],
                        'Wind':['Strong', 'Strong', 'Strong', 'Strong'],
                        'Water':['Warm', 'Cool', 'Cool', 'Warm'],
                        'Forecst':['Change', 'Same', 'Change', 'Change']})

# 라벨인코딩...
for col in new.columns.tolist():
    print(col)
    name = f'{col}_encoded'
    new[name] = le.fit_transform(new[col].values)
    
display(new)
```

### # 분류기 이용 -> prediction

```python
# 학습된 모델분류기로 새로 들어온 데이터의 Y값을 prediction한다
new['label_pred'] = clf.predict(new[x_columns].values)
# prediction의 확률도 계산해준다.
y_pred_proba = clf.predict_proba(new[x_columns].values)
print(y_pred)
print(y_pred_proba)
# pred 0일 확률
new['Prob[0]'] = y_pred_proba[:,0]
# pred 1일 확률
new['Prob[1]'] = y_pred_proba[:,1]
```

```python
# 가우시안을 썼을 때,
GNB = new.copy()
```

```python
# 베르누이를 썼을 때,
BNB = new.copy()
```

```python
# 다항분포를 썼을 때,
MNB = new.copy()
```

```python
print('가우시안 분포')
display(GNB)
print('\n\n=============================================================')
print('\n\n베르누이 분포')
display(BNB)
print('\n\n=============================================================')
print('\n\n다항 분포')
display(MNB)
```
